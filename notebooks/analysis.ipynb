```python
from river import stream
X_y = stream.iter_csv(
    dataset.path,
    converters={
        'bikes': int,
        'clouds': int,
        'humidity': int,
        'pressure': float,
        'temperature': float,
        'wind': float
    },
    parse_dates={'moment': '%Y-%m-%d %H:%M:%S'},
    target='bikes'
)
x, y = next(X_y)
x, y
```

```python
from river import stats

mean = stats.Mean()
var = stats.Var()
min_val = stats.Min()
max_val = stats.Max()
quantile_25 = stats.Quantile(0.25)
quantile_50 = stats.Quantile(0.5)
quantile_75 = stats.Quantile(0.75)

for x, y in X_y:
    mean.update(y)
    var.update(y)
    min_val.update(y)
    max_val.update(y)
    quantile_25.update(y)
    quantile_50.update(y)
    quantile_75.update(y)

print(f"Mean: {mean.get():.2f}")
print(f"Variance: {var.get():.2f}")
print(f"Minimum: {min_val.get():.2f}")
print(f"Maximum: {max_val.get():.2f}")
print(f"25th Percentile: {quantile_25.get():.2f}")
print(f"50th Percentile (Median): {quantile_50.get():.2f}")
print(f"75th Percentile: {quantile_75.get():.2f}")
```

```python
from river import compose
from river import linear_model
from river import metrics
from river import evaluate
from river import preprocessing
from river import optim

model = compose.Select('clouds', 'humidity', 'pressure', 'temperature', 'wind')
model |= preprocessing.StandardScaler()
model |= linear_model.LinearRegression(optimizer=optim.SGD(0.001))

metric = metrics.MAE()

evaluate.progressive_val_score(dataset, model, metric, print_every=20_000)
```

```python
from river import feature_extraction
from river import stats

def get_hour(x):
    x['hour'] = x['moment'].hour
    return x

model = compose.Select('clouds', 'humidity', 'pressure', 'temperature', 'wind')
model += (
    get_hour |
    feature_extraction.TargetAgg(by=['station', 'hour'], how=stats.Mean())
)
model |= preprocessing.StandardScaler()
model |= linear_model.LinearRegression(optimizer=optim.SGD(0.001))
```

```python

metric = metrics.MAE()

evaluate.progressive_val_score(dataset, model, metric, print_every=20_000)

```
